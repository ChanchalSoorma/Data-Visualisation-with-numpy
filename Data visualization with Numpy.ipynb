1. Line Plot:

- Basic Line Plot

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

x = np.linspace(0, 10, 100)      # Creating an array of 100 points from 0 to 10
y = np.cos(x)                    # Calculating the cosine of each point in the array x

plt.plot(x, y)                   # Plotting x vs y as a line plot
plt.xlabel('x-axis')             # Label for the x-axis
plt.ylabel('y-axis')             # Label for the y-axis
plt.title('Basic Line Plot')     # Title of the plot
plt.show()                       # Display the plot
```

- Custom Line Graph

```python
import matplotlib.pyplot as plt   # Importing Matplotlib for plotting

x = np.linspace(0, 10, 50)        # Creating an array of 50 points from 0 to 10
y = np.cos(x)                     # Calculating the cosine of each point in the array x

# Plotting x vs y with specified styling:
# 'o' denotes circular markers, linestyle is dashed ('--'), color is red,
# markersize is 5, and alpha (transparency) is set to 0.5

plt.plot(x, y, 'o', linestyle='--', color='red', markersize=5, alpha=0.5)

plt.xlabel('x-axis')              # Label for the x-axis
plt.ylabel('y-axis')              # Label for the y-axis
plt.title('Basic Line Plot')      # Title of the plot
plt.show()                        # Display the plot
```

- Line Plot with Multiple Lines

```python
import matplotlib.pyplot as plt   # Importing Matplotlib for plotting

x = np.linspace(0, 10, 50)        # Creating an array of 50 points from 0 to 10
y1 = np.sin(x)                    # Calculating the sine of each point in the array x
y2 = np.cos(x)                    # Calculating the cosine of each point in the array x

# Plotting y1 (sin(x)) with specified styling: red dashed line, circular markers,
# markersize 5, alpha 0.5, and label 'sin(x)'

plt.plot(x, y1, 'o', linestyle='--', color='red', markersize=5, alpha=0.5, label='sin(x)')

# Plotting y2 (cos(x)) with specified styling: blue dashed line, circular markers,
# markersize 5, alpha 0.5, and label 'cos(x)'

plt.plot(x, y2, 'o', linestyle='--', color='blue', markersize=5, alpha=0.5, label='cos(x)')

plt.xlabel('x-axis')              # Label for the x-axis
plt.ylabel('y-axis')              # Label for the y-axis
plt.title('Basic Line Plot')      # Title of the plot
plt.legend()                      # Displaying the legend to label sin(x) and cos(x)
plt.show()                        # Display the plot
```

2. Scatter Plot:

- Basic Scatter Plot

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating two arrays of 500 random numbers between 0 and 1 for x and y coordinates
x = np.random.rand(500)
y = np.random.rand(500)

# Creating a scatter plot with x and y as coordinates
plt.scatter(x, y)

# Adding labels to the x and y axes
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Adding a title to the scatter plot
plt.title('Basic Scatter Plot')

# Displaying the scatter plot
plt.show()
```

- Customized Scatter Plot

```python
import matplotlib.pyplot as plt
x = np.random.rand(500)
y = np.random.rand(500)
z = np.random.rand(500)

plt.scatter(x,y, marker = 'o', c = z, alpha = .7, cmap = 'viridis')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Scatter Plot')
plt.colorbar()
plt.show()
```

- Scatter Plot with Multiple Classes

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating two sets of 200 random points for x1, y1 in the range [-0.5, 0.5]
x1 = np.random.rand(200) - 0.5
y1 = np.random.rand(200) - 0.5

# Generating two sets of 200 random points for x2, y2 in the range [0.5, 1.5]
x2 = np.random.rand(200) + 0.5
y2 = np.random.rand(200) + 0.5

# Creating a scatter plot for the first set of points (x1, y1) with circular markers and 70% transparency
plt.scatter(x1, y1, marker='o', alpha=0.7)

# Creating a scatter plot for the second set of points (x2, y2) with circular markers and 70% transparency
plt.scatter(x2, y2, marker='o', alpha=0.7)

# Adding labels to the x and y axes
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Adding a title to the scatter plot
plt.title('Scatter Plot')

# Adding a colorbar (though not used in this case, as the scatter plot has no color mapping)
plt.colorbar()

# Displaying the scatter plot
plt.show()
```

- Hexbinning Scatter Plot

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating two arrays of 10,000 random numbers between 0 and 1 for x1 and y1
x1 = np.random.rand(10000)
y1 = np.random.rand(10000)

# Creating a hexbin plot to visualize the density of points in a 2D grid
# 'gridsize' controls the number of hexagons in the plot
plt.hexbin(x1, y1, gridsize=20)

# Adding labels to the x and y axes
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Adding a title to the hexbin plot
plt.title('Hexbining Plot')

# Adding a colorbar to show the density scale
plt.colorbar()

# Displaying the hexbin plot
plt.show()
```

3. Bar Graph:

- Basic Bar Graph

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Defining the categories (A, B, C, D) for the x-axis
x = np.array(['A', 'B', 'C', 'D'])

# Defining the corresponding values for each category on the y-axis
y = np.array([10, 20, 30, 40])

# Creating a bar graph with categories on the x-axis and values on the y-axis
plt.bar(x, y)

# Adding labels to the x and y axes
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Adding a title to the bar graph
plt.title('Basic Bar Graph')

# Displaying the bar graph
plt.show()
```

- Stacked Bar Graph

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating a 10x5 matrix with random values between 0 and 1
data = np.random.rand(10, 5)

# Plotting stacked bar charts for each dataset by adding columns progressively
# The bars represent the sum of columns in data, starting with the first column (D-1) and adding subsequent columns for each subsequent dataset (D-2 to D-5)
plt.bar(np.arange(10), data[:, 0] + data[:, 1] + data[:, 2] + data[:, 3] + data[:, 4], label='D-5')
plt.bar(np.arange(10), data[:, 0] + data[:, 1] + data[:, 2] + data[:, 3], label='D-4')
plt.bar(np.arange(10), data[:, 0] + data[:, 1] + data[:, 2], label='D-3')
plt.bar(np.arange(10), data[:, 0] + data[:, 1], label='D-2')
plt.bar(np.arange(10), data[:, 0], label='D-1')

# Adding a legend to label each dataset (D-1 to D-5)
plt.legend()

# Displaying the stacked bar chart
plt.show()
```

- Grouped Bar Graph

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Creating a 3x3 matrix (3 rows and 3 columns) with some data
data = np.array([[10, 20, 30],
                 [40, 50, 60],
                 [70, 80, 90]])

# Setting the width of the bars in the bar chart
width = 0.25

# Creating an array for the x positions of the bars
x = np.arange(3)

# Plotting the first set of bars for D-1 (shifted to the left by 'width')
plt.bar(x - width, data[:, 0], width, label='D-1')

# Plotting the second set of bars for D-2 (centered on x positions)
plt.bar(x, data[:, 1], width, label='D-2')

# Plotting the third set of bars for D-3 (shifted to the right by 'width')
plt.bar(x + width, data[:, 2], width, label='D-3')

# Adding the legend to label each data set
plt.legend()

# Displaying the bar chart
plt.show()
```

4. Histogram:

- Basic Histogram

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating a dataset of 10,000,000 random numbers from a standard normal distribution
data = np.random.randn(10000000)

# Plotting the histogram of the data with 1000 bins
plt.hist(data, bins=1000)

# Displaying the histogram plot
plt.show()
```

- Cumulative Histogram

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating a dataset of 1,000,000 random numbers from a standard normal distribution
data = np.random.randn(1000000)

# Plotting a cumulative histogram with 1000 bins
# The 'cumulative=True' argument makes the histogram cumulative
plt.hist(data, bins=1000, cumulative=True)

# Displaying the cumulative histogram plot
plt.show()
```

- Histogram with Multiple Datasets

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating two datasets of 100,000 random numbers from a standard normal distribution
# data2 is shifted by 2 units compared to data1
data1 = np.random.randn(100000)
data2 = np.random.randn(100000) + 2

# Plotting the histogram for data1 with 100 bins and an alpha (transparency) value of 0.7
# 'label' is used to provide a legend label for data1
plt.hist(data1, bins=100, alpha=0.7, label='D1')

# Plotting the histogram for data2 with 100 bins and an alpha (transparency) value of 0.7
# 'label' is used to provide a legend label for data2
plt.hist(data2, bins=100, alpha=0.7, label='D2')

# Displaying the legend for the plot to distinguish between D1 and D2
plt.legend()

# Displaying the histogram plot
plt.show()
```

5. Box Plot:

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating a 100x5 matrix of random numbers from a standard normal distribution
data = np.random.randn(100, 5)

# Creating a box plot for the data with custom labels for each dataset (D-1 to D-5)
# 'vert=False' makes the box plot horizontal
plt.boxplot(data, labels=['D-1', 'D-2', 'D-3', 'D-4', 'D-5'], vert=False)

# Displaying the plot
plt.show()
```

6. Heatmap:

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Generating a 10x10 matrix of random numbers between 0 and 1
data = np.random.rand(10, 10)

# Displaying the data as an image (matrix of values) using imshow()
plt.imshow(data)

# Adding a color bar to show the mapping of values to colors
plt.colorbar()

# Displaying the image with the color bar
plt.show()
```

7. Pie Chart:

```python
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Defining the labels for the pie chart (programming languages)
labels = ['Python', 'C++', 'Java', 'JavaScript', 'PHP']

# Defining the corresponding popularity sizes for each language
sizes = [21.3, 14.6, 9.2, 8.5, 7.7]

# Defining the explode values for each slice (to offset certain slices for emphasis)
explode = (0.13, 0.1, 0, 0, 0)

# Defining custom colors for each slice of the pie
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']

# Plotting the pie chart with labels, colors, and percentages displayed
plt.title("Programming Language Popularity")  # Title for the pie chart
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.0f%%', shadow=True, explode=explode, startangle=140)

plt.axis('equal')  # Ensuring the pie chart is drawn as a circle (equal aspect ratio)
plt.legend(title="Languages", loc="upper left")  # Adding a legend to label each programming language

plt.show()  # Displaying the pie
```
